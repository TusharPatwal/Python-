## Merge, Join, and Concatenate 

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Tushar\Downloads\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Tushar\Downloads\LOTR 2.csv")
df2
```

### Merge

```python
# (how='inner') it is by default
df1.merge(df2, how='inner', on='FellowshipID')
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
```

### Joins

```python
df1.join(df2, how='outer', on='FellowshipID',lsuffix='_Left', rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),how='outer',lsuffix='_Left', rsuffix='_Right')
df4
```

### Concatenate

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2], join='outer')
```

```python
pd.concat([df1,df2], join='outer',axis=1)
```

### Apppned

```python
df1.append(df2)
```
