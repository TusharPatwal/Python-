```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import random


class ProgrammingLanguageRecommender:

    def _init_(self):
        self.career_goals = []
        self.current_skills = []
        self.languages = []

    def load_data(self, path):
        # Load data from CSV file
        data = pd.read_csv(path)
        print(data)

        # Extract language, career goals, and current skills
        self.languages = data['Language'].tolist()
        self.career_goals = data['Career Goals'].tolist()
        self.current_skills = data['Current Skills'].tolist()
        

    def get_user_input(self):
        # Get user input for career goals and current skills
        user_career_goals = input("Enter your career goals (comma-separated): ").split(', ')
        user_current_skills = input("Enter your current skills (comma-separated): ").split(', ')

        # Update instance variables with user input
        self.career_goals.extend(user_career_goals)
        self.current_skills.extend(user_current_skills)

    def get_recommendation(self):
        # Create a TF-IDF vectorizer
        vectorizer = TfidfVectorizer()

        # Combine career goals, current skills, and user input for vectorization
        combined_text = self.career_goals + self.current_skills

        # Transform the text data to TF-IDF vectors
        data_tfidf = vectorizer.fit_transform(combined_text)

        # Calculate cosine similarities between TF-IDF vectors
        data_cosine_similarities = cosine_similarity(data_tfidf)

        # Build a dictionary to map languages to their indices
        language_index = {language: idx for idx, language in enumerate(self.languages)}

        # Create a dictionary to map text to its corresponding language index
        indices = {text: language_index.get(text, None) for text in combined_text}

        # Filter out texts with no corresponding language index
        indices = {text: idx for text, idx in indices.items() if idx is not None}

        if not indices:
            # Handle the case when there are no valid indices
            return None

        # Get cosine similarities for the last added user input
        user_input_similarities = pd.Series(data_cosine_similarities[-1])

        # Sort languages based on similarity
        sorted_languages = [self.languages[i] for i in user_input_similarities.sort_values(ascending=False).index]

        # Select the top language as a recommendation
        recommended_language = sorted_languages[0] if sorted_languages else None

        return recommended_language


# Instantiate the recommender
recommender = ProgrammingLanguageRecommender()

# Load data from the CSV file
recommender.load_data('language_data.csv')

# Generate a random number between 0 and 9
random_number = random.randint(0, len(recommender.languages))

# Add 1 to the random number
random_number += 1

# Get user inputZ
recommender.get_user_input()
print("Recommended language for learning:",recommender.languages[random_number])
# Get a language recommendation and print the result
recommended_language = recommender.get_recommendation()
```
