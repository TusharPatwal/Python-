```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

Line Chart
===
A line chart is a type of chart used to show information that changes over time.

```python
# To draw a line chart, we use the plt.plot function
yield_apples = [.895,.91,.919,.926,.929,.931]
```

```python
plt.plot(yield_apples)
```

```python
# We can include a semicolon(;) at the end of the last statement in the cell to avoiding the output and just display the graph
plt.plot(yield_apples);
```

Costomizing the X-axis
---
The X-axis Of the plot shows list element indexes 0 to 5. The plot would be more informative if we could show the year for which the data is being plotted. We can do this by two argument 
`plt.plot`.

```python
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [.895,.91,.919,.926,.929,.931]
```

```python
plt.plot(years,yield_apples)
```

Axis Labels
---
We can add labels to axis to show what each asix is represents using the `plt.xlabel` and `plt.ylabel` methods.

```python
plt.plot(years,yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

Plotting Multiple Lines
---
It's really easy to plot multiple lights in same graph. Just invoke `plt.plot` function multiple times. Let's compare the yields of apples vs. oranges in Kanto.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

Chart Title and Legend
---
To differentiate between multiple lines, we can include a legend within the graph using `plt.legend` function. We also give the entire chart a title using `plt.title` function.

```python
plt.plot(years,apples)
plt.plot(years,oranges) 

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

Line Markers
---
We can show markers for the data points on each line using the `marker` argument of `plt.plot`. Matplotlib supports many different types of markers like circle, cross, square, dimond etc.

```python
plt.plot(years,apples, marker='o')
plt.plot(years,oranges, marker='x') 

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

Styling lines and markers
---
The `plt.plot` function support menu argument for styling lines and markers:
- `color` or `c`: Set the color of line.
- `linestyle` or `ls`: choose between a solid or dashed line
- `linewidth` or `lw`: set the width of a line 
- `markersize` or `ms`: set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width of markers
- `markerfacecolour` or `mfc`: Set the fill color for markers
- `alpha`: opacity of the plot

```python
plt.plot(years,apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years,oranges, marker='o', c='r', ls='--', lw=3, ms=10,alpha=.5) 

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

The `fmt` argument provides a shorthand for speifying the line style, marker and line color. It can be provided as third argument to `plt.plot`.

        fmt = '[marker][line][color]'

```python
plt.plot(years,apples, 's-r')
plt.plot(years,oranges, 'o--', c='orange') 

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

Changing the Figure Size
---
You can use the `plt.figure` function to change the size of the figure.

```python
plt.figure(figsize=(6,4))

plt.plot(years,oranges,'or')
plt.title('Yield (tons per hectare)');
```

Scatter Plot
===
In a scatter plot, the values of 2 variables are plotted as points on a a 2-dimentional grid. Additionally, you can also use a third variable to determine the size or color of the points. Let's try out an example.

```python
# load data into a Pandas dataframe
flower_df = pd.read_csv('Iris.csv')
```

```python
del flower_df['Id']
```

```python
flower_df
```

```python
flower_df.Species.unique()
```

```python
plt.plot(flower_df.SepalLengthCm, flower_df.SepalWidthCm);
```

We can use a scatter plot to visualize how sepal length & sepal width vary using the `scatterplot` function from `seaborn`(import as `sns`)

```python
sns.scatterplot((flower_df.SepalLengthCm, flower_df.SepalWidthCm));
```

```python
sns.scatterplot(((flower_df.SepalLengthCm, flower_df.SepalWidthCm, flower_df.Species)));
```
