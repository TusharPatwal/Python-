Scatter plot
---

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset('penguins').head(20)
```

```python
var.head()
```

```python
var.columns
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=var,hue='sex');
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=var,hue='sex',style='sex',size='sex',sizes=(70,30),palette='gist_ncar');
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',
                data=var,hue='sex',style='sex',size='sex',
                sizes=(70,30),palette='gist_ncar',alpha=1);
```
