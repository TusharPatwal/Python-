Line Plot
---

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var1 = [2,3,4,5,6,8,9]
x = pd.DataFrame({"var":var,"var1":var1})

# seaborn
sns.lineplot(x='var',y='var1', data=x)
plt.show()
```

```python
# matplotlib
plt.plot(var,var1)
plt.show()
```

```python
# how to get built in datasets
y_1 = sns.load_dataset('penguins').head(20)
```

```python
y_1
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y_1,hue='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y_1,hue='sex',size=15)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y_1,hue='sex', style='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y_1,hue='sex', style='sex',palette='ocean_r',markers=['o','>'],
            dashes=True, legend=True)
plt.grid()
plt.title('Python')
plt.show()
```
